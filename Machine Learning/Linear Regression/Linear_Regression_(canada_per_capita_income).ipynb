```python
# importing necessary libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
df.shape
```

```python
df_year = df.drop('per capita income (US$)',axis='columns')
df_year.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='+')
```

```python
#Creating Linear Regression Object
reg = linear_model.LinearRegression()
reg.fit(df_year,df['per capita income (US$)'])
```

## (1) Predict Canada's per capita income in year 2020


```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('price')
plt.scatter(df.year,df['per capita income (US$)'],marker='+')
plt.plot(df.year,reg.predict(df_year),color='red')
```
