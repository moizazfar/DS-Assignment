```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
dataset = df.drop(["PassengerId","SibSp","Parch","Ticket","Fare","Cabin","Embarked"],axis="columns")
```

```python
#Filling missing values by mean
dataset['Age'].fillna(dataset['Age'].mean(),inplace=True)
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.dtypes
```

```python
dataset['Age'] = dataset['Age'].astype(int)
```

```python
dataset.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
name_n = LabelEncoder()
gender_n = LabelEncoder()
```

```python
le_name = name_n.fit_transform(dataset['Name'])
le_gender = gender_n.fit_transform(dataset['Sex'])
```

```python
dataset['name_n']=le_name
dataset['gender_n']=le_gender
```

```python
dataset
```

```python
encoded_data = dataset.drop(['Name','Sex'],axis='columns')
```

```python
encoded_data
```

```python
X = encoded_data.drop('Survived',axis='columns').values
```

```python
X
```

```python
y=encoded_data.Survived.values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
X_test
```

```python
#Importing Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
print("Accuracy = ", model.score(X_test, y_test)*100)
```

```python
print("Predict value " + str(model.predict([X[5]])))
print("Real value " + str(y[5]))
```
